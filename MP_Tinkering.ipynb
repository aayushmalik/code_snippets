```python
import pandas as pd
import numpy as np
```

```python
raw_df = pd.read_excel('MP_Data_2017.xlsx')
```

```python
raw_df.head(10)
```

```python
df  = pd.DataFrame()
```

```python
df['district'] = raw_df['District']
```

```python
df = pd.concat([df]*53, ignore_index=True)
```

```python
df.head(20)
```

```python
df.shape
```

```python
area_vector = []
production_vector = []
```

```python
for i in range(1, 107, 2):
    temp_area = raw_df.iloc[:,i]
    temp_production = raw_df.iloc[:,i+1]
    area_vector.extend(temp_area)
    production_vector.extend(temp_production)
```

```python
len(area_vector)
```

```python
area_vector
```

```python
df['area'] = area_vector
```

```python
df['production'] = production_vector
```

```python
df.head(10)
```

```python
columns = []
for column in raw_df.columns:
    columns.append(column)
    
print(columns)
```

```python
type(columns)
```

```python
crops = []
for i in range(1, 107, 2):
    crops.append(columns[i])

print(crops)
```

```python
len(crops)
```

```python
df.sample(10)
```

```python
df['crop'] = 0
```

```python
for i in range(0, 53):
    lower_bound = 51*i
    upper_bound = 51*(i) + 1 #inclusive so in the formula don't forget to add 1
    df.iloc[lower_bound:upper_bound, 3] = crops[i]
```

```python
df.to_excel('MP_Cleaned_Excel.xlsx', index = None)
```
