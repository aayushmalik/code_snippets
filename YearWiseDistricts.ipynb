```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('cropdata19972017.xlsx')
```

```python
df.head(10)
```

```python
#yearwisedf = pd.DataFrame()
yearwisedf = pd.DataFrame(index=range(800))

def processing(df):
    for year in range(1997, 2017):
        yearwisedistricts = df[df.YEAR == year]['DISTNAME'].unique()
        yearwisedf[year] = pd.Series(yearwisedistricts)
```

```python
processing(df)
```
