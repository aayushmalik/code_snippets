```python
def data_extraction(file_path):
    '''
    this function takes the file path from the directory and then performs the reiterable operations. 
    the output is a dataframe which is appended to the master dataframe
    '''
    data = pd.read_html(file_path)
    df = data[0]
    malaria = df.filter(regex=VARIABLE)
    values = malaria.iloc[1:,0]
    districts = df.iloc[1:,1]
    final_output = pd.DataFrame(data={'district':districts, '%malaria_children_under_5':values})
    return final_output
```

```python
import glob
```

```python
final_output_data = []
```

```python
for filepath in glob.iglob(r'C:\Users\MalikAayush\Desktop\HarvardWork\HMIS_Data\2015_2016Data\*.xls'):
    print(filepath)
    output_df = data_extraction(filepath)
    final_output_data.append(output_df)
```

```python
final_output_data
```

```python
df = pd.concat(final_output_data)
```
