```python
import time

import re

from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('clean_locations.csv')
```

```python
def find_coordinates(location):
    '''this function takes the input value of a location and returns its coordinates'''

    driver = webdriver.Chrome(r'C:\Users\Arvind Balyan\Desktop\AayushTCD\chromedriver_win32\chromedriver.exe')  # Optional argument, if not specified will search path.
    driver.get('https://www.google.com/maps');
    search_box = driver.find_element_by_name('q')
    search_box.send_keys(location)
    driver.find_element_by_css_selector("button[id='searchbox-searchbutton']").click()
    WebDriverWait(driver, 10).until(EC.url_contains("https://www.google.com/maps/place/"))
    time.sleep(15)
    coordinates = driver.current_url
    driver.quit()
    coordinates_list = re.findall(r"\@(.*?)\/", coordinates)
    return coordinates_list[0].split(',')[0], coordinates_list[0].split(',')[1]
```

```python
for index, row in df.iterrows():
    try:
        if np.isnan(float(row['gpslat'])) == True:
            lat, long = find_coordinates(row['location_clean'])
            print(lat,long)
            df.at[index, 'gpslat'] = lat
            df.at[index, 'gpslong'] = long
    except:
        print('Error Encountered at', row['location_clean'], 'in row', index, 'Moving to NEXT')
        continue
```

```python
df.to_csv('geocoordinatesfile.csv', index = None, header = True)
```
