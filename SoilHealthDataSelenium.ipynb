```python
import pandas as pd
import numpy as np
import time
from selenium import webdriver
```

```python
states = ['Andaman And Nicobar Islands', 'Andhra Pradesh', 'Arunachal Pradesh', 'Assam', 'Bihar', 'Chandigarh', 'Chhattisgarh', 'Dadra And Nagar Haveli', 'Daman And Diu', 'Delhi', 'Goa', 'Gujarat', 'Haryana', 'Himachal Pradesh', 'Jammu And Kashmir', 'Jharkhand', 'Karnataka', 'Kerala', 'Ladakh', 'Lakshadweep', 'Madhya Pradesh', 'Maharashtra', 'Manipur', 'Meghalaya', 'Mizoram', 'Nagaland', 'Odisha', 'Puducherry', 'Punjab', 'Rajasthan', 'Sikkim', 'Tamil Nadu', 'Telangana', 'Tripura', 'Uttar Pradesh', 'Uttarakhand', 'West Bengal']
```

```python
nutrients = ["Nitrogen(N)", "Phosphorous(P)", "Potassium(K)", "Organic Carbon(OC)"]
```

```python
for nutrient in nutrients:
    for state in states:
        print(state)
        PATH = "C:\Program Files (x86)\chromedriver.exe"
        driver = webdriver.Chrome(PATH)

        driver.get("https://soilhealth.dac.gov.in/PublicReports/DistrictMacroNSPercentagewise")

        #the following is done to remove the language selection division from the page, else it interferes with the code for downloading files later
        element = driver.find_element_by_xpath("//body[@class='BackGroundNew']/div[@id='body']/div/div[1]")
        driver.execute_script("""
        var element = arguments[0];
        element.parentNode.removeChild(element);
        """, element)
        time.sleep(3)

        driver.find_element_by_xpath("//select[@id='State_Code']").click()
        time.sleep(0.5)
        driver.find_element_by_xpath("//option[contains(text(),'"+state+"')]").click()
        time.sleep(0.5)
        driver.find_element_by_xpath("//select[@id='CycleId']").click()
        time.sleep(0.5)
        driver.find_element_by_xpath("//option[contains(text(),'2015-16 to 2016-17')]").click()
        time.sleep(0.5)
        driver.find_element_by_xpath("//select[@id='Nutrient']").click()
        time.sleep(0.5)
        driver.find_element_by_xpath("//option[contains(text(),'"+nutrient+"')]").click()
        time.sleep(0.5)
        driver.find_element_by_xpath("//input[@id='confirmLink']").click()
        time.sleep(10)

        iframe = driver.find_element_by_xpath("//div[@id='report']//iframe")
        driver.switch_to.frame(iframe)

        time.sleep(1)

        driver.find_element_by_id("ReportViewer1_ctl05_ctl04_ctl00_ButtonLink").click()
        time.sleep(1)
        driver.find_element_by_xpath("//a[contains(text(),'Excel')]").click()
        time.sleep(2)

        driver.close()
```
