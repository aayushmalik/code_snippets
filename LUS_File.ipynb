```python
import pandas as pd
import numpy as np
import time
from selenium import webdriver
```

```python
states = ['Andaman and Nicobar Islands', 'Andhra Pradesh', 'Arunachal Pradesh', 'Assam', 'Bihar', 
          'Chandigarh', 'Chhattisgarh', 'Dadra and Nagar Haveli', 'Daman and Diu', 'Delhi', 'Goa', 'Gujarat', 
          'Haryana', 'Himachal Pradesh', 'Jammu and Kashmir ',
'Jharkhand',
'Karnataka',
'Kerala',
'Lakshadweep',
'Madhya Pradesh',
'Maharashtra',
'Manipur',
'Meghalaya',
'Mizoram',
'Nagaland',
'Odisha',
'Puducherry',
'Punjab',
'Rajasthan',
'Sikkim',
'Tamil Nadu',
'Telangana ',
'Tripura',
'Uttar Pradesh',
'Uttarakhand',
'West Bengal']
```

```python
years = ['2015-16', '2016-17', '2017-18']
```

```python
for year in years:
    for state in states:
        try:
            print(year, state)
            PATH = "C:\Program Files (x86)\chromedriver.exe"
            driver = webdriver.Chrome(PATH)

            driver.get("https://aps.dac.gov.in/LUS/Public/Reports.aspx")
            print(driver.title)
            
            driver.find_element_by_xpath("//select[@name='DdlState']").click()
            time.sleep(0.5)
            driver.find_element_by_xpath(f"//option[contains(text(),'{state}')]").click()
            time.sleep(0.5)
            driver.find_element_by_xpath("//select[@name='DdlYear']").click()
            time.sleep(0.5)
            driver.find_element_by_xpath(f"//option[contains(text(),'{year}')]").click()
            time.sleep(0.5)
            driver.find_element_by_xpath("//select[@name='DdlFormat']").click()
            time.sleep(0.5)
            driver.find_element_by_xpath("//option[contains(text(),'Excel')]").click()
            time.sleep(0.5)
            
            driver.find_element_by_xpath("//a[contains(text(),'1. Classification of Area')]").click()
            time.sleep(10)

            driver.close()
            print(f'Done for {state} and {year}')
        except:
            print(f'Error encountered for year {year} and state {state}')
            pass
```
