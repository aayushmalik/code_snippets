```python
import pandas as pd
import numpy as np
import os

def parse_file(file):
    df = pd.read_excel(file)
    df = df[df['YEAR'] == 1966]

    cols_to_drop = 'STCODE STNAME DIST YEAR'.split()
    df = df.drop(cols_to_drop, axis=1).reset_index().drop('index', axis=1)

    df = df.rename(columns={
            'RICE_TA':'Rice area',
            'RICE_TQ':'Rice production',
            'WHT_TA':'Wheat area',
            'WHT_TQ':'Wheat production',
            'SORG_KA':'Kharif sorghum area',
            'SORG_KQ':'Kharif sorghum production',
            'SORG_RA':'Rabi sorghum area',
            'SORG_RQ':'Rabi sorghum production',
            'SORG_TA':'Total sorghum area',
            'SORG_TQ':'Total sorghum production',
            'PMLT_TA':'Pearl millet area',
            'PMLT_TQ':'Pearl millet production',
            'MAIZ_TA':'Maize area',
            'MAIZ_TQ':'Maize production',
            'FMLT_TA':'Finger millet area',
            'FMLT_TQ':'Finger millet production',
            'BRLY_TA':'Barley area',
            'BRLY_TQ':'Barley production',
            'CERL_TA':'Total cereals area',
            'CERL_TQ':'Total cereals production',
            'CPEA_TA':'Chickpea area',
            'CPEA_TQ':'Chickpea production',
            'PPEA_TA':'Pigeonpea area',
            'PPEA_TQ':'Pigeonpea production',
            'MPUL_TA':'Minor pulses area (includes all pulses other than chickpea and pigeonpea)',
            'PULS_TA':'Total pulses area',
            'PULS_TQ':'Total pulses production',
            'GNUT_TA':'Groundnut area',
            'GNUT_TQ':'Groundnut production',
            'SESA_TA':'Sesamum area',
            'SESA_TQ':'Sesamum production',
            'RM_TA':'Rape and mustard area',
            'RM_TQ':'Rape and mustard production',
            'SAFF_TA':'Safflower area',
            'SAFF_TQ':'Safflower production',
            'CAST_TA':'Castor area',
            'CAST_TQ':'Castor production',
            'LINS_TA':'Linseed area',
            'LINS_TQ':'Linseed production',
            'SUNF_TA':'Sunflower area',
            'SUNF_TQ':'Sunflower production',
            'SOYA_TA':'Soyabean area',
            'SOYA_TQ':'Soyabean production',
            'OILS_TA':'Total oil seeds area',
            'OILS_TQ':'Total oil seeds production',
            'SCAN_TA':'Sugarcane area',
            'SGUR_TQ':'Sugarcane (gur) production',
            'COTN_TA':'Cotton area',
            'COTN_TQ':'Cotton (lint) production',
            'FRUT_TA':'Fruits area',
            'VEGT_TA':'Vegetables area',
            'TFV_TA':'Total fruits and vegetables area',
            'POTA_TA':'Potatoes area',
            'ONIO_TA':'Onions area',
        })

    df = pd.melt(df,id_vars=['DISTNAME'])

    df['area']=0
    df['production']=0

    for i, row in df.iterrows():
        if 'area' in row.variable:
            df.iloc[i,3] = row['value']
        else:
            df.iloc[i-len(df.DISTNAME.unique()), 4] = row['value']

    word = "production"
    df = df[~df["variable"].str.contains(word)]
    df.variable = df.variable.replace({'area':''}, regex=True)

    df['yield'] = df['production']/df['area']
    df = df.drop('value', axis =1)
    df = df.rename(columns={
        'variable':'crop'
    })
    return df

master_df = pd.DataFrame().fillna(0)

for file in os.listdir(r'C:\Users\AayushMalik\Desktop\HarvardWork\state database'):
    df = parse_file(r'C:\Users\AayushMalik\Desktop\HarvardWork\state database'+'\\'+file)
    master_df = master_df.append(df, ignore_index = True)

master_df = master_df.replace(master_df[master_df._get_numeric_data()<0], '')
master_df.to_csv('master_data_file_VDSA.csv', index=None)
```
